```python
import matplotlib.pyplot as plt

# Data for the models and their accuracy (results are stored in the file 'Models.ipynb')
models = ['Baseline model (NB)', 'RF default', 'RF after SMOTE data balancing', 'RF after feature selection','RF after optimisation']
accuracy = [52.93, 67.95, 74.21, 74.08, 75.47]

# Creating the line graph with updates
plt.figure(figsize=(10, 5))
plt.plot(models, accuracy, marker='o', color='orange', linewidth=2)  # Setting the line color to yellow
plt.title('Evolution of Random Forest after Applying Different Techniques')
plt.ylabel('Accuracy (%)')
plt.ylim(40, 80)
plt.grid(True)
plt.xticks(rotation=45)

# Adding data labels to each point
for i, acc in enumerate(accuracy):
    plt.text(models[i], acc + 1, f'{acc}%', ha='center')

plt.show()
```
